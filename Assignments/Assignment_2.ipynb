### Translate the image towards right by 400 and up by 150 pixels.

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = plt.imread('../images/sachin.jpg')
```

```python
plt.title("Original Image")
plt.imshow(img)
plt.axis('off')
```

```python
# Get the dimensions of the image
height, width = img.shape[:2]

# Define the translation matrix
translate_x = 400  # pixels to the right
translate_y = -150  # pixels up
translation_matrix = np.float32([[1, 0, translate_x], [0, 1, translate_y]])

# Perform the translation
translated_img = cv2.warpAffine(img, translation_matrix, (width + translate_x, height + abs(translate_y)))

## Convert BGR image (OpenCV default) to RGB (Matplotlib default)
# translated_img_rgb = cv2.cvtColor(translated_img, cv2.COLOR_BGR2RGB)
```

```python
plt.title("Translated Image")
plt.imshow(translated_img)
plt.axis('off')
plt.show()
```
